```python
import time

# Implementacin usando la solucin matemtica directa
def solucion_directa(n):
    return (1 + n) * (2 ** n)

# Implementacin usando la forma recursiva
def solucion_recursiva(n):
    if n == 0:
        return 1
    elif n == 1:
        return 4
    else:
        return 4 * solucion_recursiva(n - 1) - 4 * solucion_recursiva(n - 2)

# Pruebas para comparar tiempos
n = 50

# Medicin de tiempo para la solucin directa
start_time = time.time()
resultado_directo = solucion_directa(n)
end_time = time.time()
print(f"Solucin directa para n={n}: {resultado_directo} (Tiempo: {end_time - start_time:.10f} segundos)")

# Medicin de tiempo para la solucin recursiva
start_time = time.time()
resultado_recursivo = solucion_recursiva(n)
end_time = time.time()
print(f"Solucin recursiva para n={n}: {resultado_recursivo} (Tiempo: {end_time - start_time:.10f} segundos)")
```
