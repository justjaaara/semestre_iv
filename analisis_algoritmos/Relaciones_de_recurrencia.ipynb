```python
import time

# Implementación usando la solución matemática directa
def solucion_directa(n):
    return (1 + n) * (2 ** n)

# Implementación usando la forma recursiva
def solucion_recursiva(n):
    if n == 0:
        return 1
    elif n == 1:
        return 4
    else:
        return 4 * solucion_recursiva(n - 1) - 4 * solucion_recursiva(n - 2)

# Pruebas para comparar tiempos
n = 50

# Medición de tiempo para la solución directa
start_time = time.time()
resultado_directo = solucion_directa(n)
end_time = time.time()
print(f"Solución directa para n={n}: {resultado_directo} (Tiempo: {end_time - start_time:.10f} segundos)")

# Medición de tiempo para la solución recursiva
start_time = time.time()
resultado_recursivo = solucion_recursiva(n)
end_time = time.time()
print(f"Solución recursiva para n={n}: {resultado_recursivo} (Tiempo: {end_time - start_time:.10f} segundos)")
```
